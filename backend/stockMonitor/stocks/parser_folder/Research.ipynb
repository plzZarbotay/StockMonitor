```python
import requests
import pandas as pd
```

```python
base_url = 'https://iss.moex.com/iss/securities.json'
params = {'start': 0, 'engine': 'stock', 'market': 'shares'}
data = []
while True:
    response = requests.get(base_url, params=params)
    if not response.json()['securities']['data']:
        break
    response = response.json()
    data += [{k : r[i] for i, k in enumerate(response['securities']['columns'])} for r in response['securities']['data']]
    params['start'] += 100
companies = pd.DataFrame(data)
```
